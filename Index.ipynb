### Условие: 

Дана матрица $n \times m$  
Матрицу можно заполнять двумя способами:  

                                                 Первый (слева направо)  
\begin{matrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{matrix}

                                                 Второй (сверху вниз)  
\begin{matrix}
1 & 4 \\
2 & 5 \\
3 & 6
\end{matrix}

Требуется найти количество таких ячеек в матрице, в которых при любом из двух способов заполнения окажется одинаковое число.  
Для примера, в показаной выше матрице $3 \times 2$ ответ будет $2$. Совпадать будет **левая верхняя** и **правая нижняя** ячейки

Для начала, научимся определять число в заданной ячейке при обоих способах заполнения. Мы имеем **номер строки $i$ и номер столбца $j$** (в нулевой индексации)  
$a = i \times m + j + 1$  
$b = j \times n + i + 1$  
$a$ - первый способ заполнения, $b$ - второй  
Как работают эти формулы?  
$i \times m + 1$ - первое число в $i$-ой строке. Добавив $j$, получаем номер в нужной нам ячейке **при первом способе заполнения**.  
Просто поменяв местами $i$ и $j$, получаем номер в нужной ячейке **при втором способе заполнения**

Исходя из условия, нужные нам ячейки имеют одинаковое значение при любом из двух заполнений. Запишем уравнение:  
$a = b$  
$i \cdot m + j + 1 = j \cdot n + i + 1$  
Выразим из этого уравнения $j$:  
$i \cdot m + j = j \cdot n + i$ (отнимаем единички, так как они есть в обоих частях уравнения)  
$j = j \cdot n + i - i \cdot m$  
$j - j \cdot n = i - i \cdot m$  
$j \cdot (1 - n) = i - i \cdot m$  
$j = \frac{i - i \cdot m}{n - 1}$

Теперь, зная номер строки $i$, можно по такой формуле узнать номер столбца $j$, на пересечении с которым находится подходящая нам ячейка. Разумеется, **не в каждой строке есть подходящая ячейка**. В таком случае результат формулы будет **вещественным числом**, а значит и не сможет стать индексом.

Так как числитель в формуле - это $i - i \cdot m$, то при вычислении если $m = 1$, то $j$ будет равен $0$, а если больше, то число будет отрицательное. По-этому нужно вывести случаи $n = 1$ и $m = 1$ в отдельный **if**, а формулу взять в **модуль**.

В итоге, получаем такую формулу: $j = |\frac{i - i \cdot m}{n - 1}|$, которая работает на всех $1 < n, m$

Теперь напишем код!

```python
n, m = map(int, input().split())
ans = 0
if n == 1 or m == 1: # выделим случай n == 1 и m == 1
    print(max(n, m)) # ответ будет максимумом из n и m. Если n == 1 и m == 1, то ответ 1 всё равно будет правильным
    exit()
for i in range(n): # переберем все строки
    x = abs((i - i * m) / (n - 1)) # и применим для каждой вышеупомянутую формулу
    if x == int(x): # проверяем число на целостность
        ans += 1 # если формула выдала целое число, значит в этой строке есть нужная ячейка, добавляем в ответ единичку
print(ans)
```
